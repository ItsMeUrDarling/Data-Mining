### About this dataset & Variable description

The data 'kc_house_data.csv' was downloaded from https://www.kaggle.com/datasets/shivachandel/kc-house-data. 

The 'price' refers to the price of the house in dollars, 'bedrooms' refers to the number of bedrooms contained in a house, 'bathrooms' refer to the number of bathrooms included in a house, 'yr_built' refers to the year the house was built, 'sqft_living' refers to the living area of the house in square feet, and 'grade' refers to a score given according to the quality of the house.


```python
# Loading necessary library 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the csv file

df = pd.read_csv ('kc_house_data.csv')
```

I obtained the mean, three quartiles, median, mean, and maximum of all the columns in the data. Mean is mainly used when the data is approximately normally distributed.

I used graphs such as scatter plots to find suspicious data, heat plots to conduct correlation analysis, and boxplots to check for outliers to clean the data. These techniques are employed to investigate whether there is a relationship between one dependent variable and several independent variables.

```python
# Scatter plot visualization

from pandas.plotting import scatter_matrix
attribute=['sqft_living15','sqft_above','sqft_living','price']
scatter_matrix(df[attribute], figsize=(12, 8))
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=df,x='condition',y='price')
plt.show()
```

The images above were made to find outliers in the dependent variable 'price', using multiple scatter plots and a boxplot. From the generated scatter plots and boxplot, we could determine that there's a significantly large number of outliers contained in the dependent variable. However, the outliers in this dataset were acceptable even if they were extreme since there is the possibility of such observations. The outliers were not deleted but kept as part of the data.

```python
#Heatplot
corr = df.corr()
sns.heatmap(corr, xticklabels = corr.columns, yticklabels = corr.columns)
```

The correlation analysis was conducted to establish the relationship between variables. The variables obtained from the data entailed 18 attributes. The analysis results show that the price of the house and square foot had a strong positive correlation (r ≈ 0.7). The relationship between other variables exhibited a very weak relationship. Their correlation coefficient was lower than 0.3.
